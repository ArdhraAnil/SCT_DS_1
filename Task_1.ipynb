```python
import pandas as pd
df=pd.read_csv("/content/1.1 - India_Historical_Population_Density_Data.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(df["year"],df["popu1ation_growth_rate"],color="blue")
plt.xlabel("year")
plt.ylabel("population")
plt.title("population density")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.hist(df["popu1ation_growth_rate"],bins=10,color="blue",edgecolor='black')
plt.xlabel("population density")
plt.ylabel("frequency")
plt.title("distribution of population density")
plt.show()
```
